```python
import pandas as pd
import requests
import bs4

targetUrl = "https://www.samsung.com/sec/galaxy/home/" #수집하려는 주소

resp = requests.get(targetUrl)

htmlnatureCode  = resp.text  #html 전체를 변수로 지정.

bs = bs4.BeautifulSoup(htmlnatureCode, "html.parser") #라이브러리를 활용해 변수 bs에 담는다.

targetDivTag = bs.find(name="div", attrs = {"class":"footer-content"}) # div태그만 찾아서 변수로 지정

ulTag = targetDivTag.find(name="ul") #ul태그만 찾아서 변수 지정.

ulTag2 = ulTag.findAll(name="ul") #위의  ul태그 아래의 ul태그만 찾아서 변수 지정. 두번째라는 뜻으로 ulTag2로 함

ulTag2_1 = ulTag2[0] #필요한 ul태그중 첫번째 리스트를 ulTag2_1로 지정

ulTag2_2 = ulTag2[1] #필요한 ul태그중 첫번째 리스트를 ulTag2_1로 지정

aTag1 = ulTag2_1.findAll(name="a") #필요한 a태그를 모아둠 (첫 번째 리스트)

aTag2 = ulTag2_2.findAll(name="a") #필요한 a태그를 모아둠 (두 번째 리스트)

aTagLen1 = len(aTag) #첫 번째 리스트 길이를 변수로 지정

aTagLen2 = len(aTag) #두 번째 리스트 길이를 변수로 지정 2

nameList = []          #빈 리스트를 지정함
linkList = []

for i in range(0,aTagLen1):          #첫번째 리스트를 nameList와 linkList에 담기 위한 for문
    tagTextName = aTag1[i].text
    tagLinkInfo = aTag1[i].attrs["href"]
    tagNameinfo = aTag1[i].name
    nameList.append(tagTextName)
    linkList.append(tagLinkInfo)

for i in range(0,aTagLen2):           #두번째 리스트를 nameList와 linkList에 담기 위한 for문
    tagTextName = aTag2[i].text
    tagLinkInfo = aTag2[i].attrs["href"]
    tagNameinfo = aTag2[i].name
    nameList.append(tagTextName)
    linkList.append(tagLinkInfo)

nameList                            #리스트 내용 확인 1

linkList                            #리스트 내용 확인 2

samsungDf = pd.DataFrame(zip(nameList,linkList), columns = ["제품","링크"])

samsungDf.to_csv("./practice1.csv",
                 index=False)
```
